```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
pokemon = pd.read_csv('pokemon.csv', error_bad_lines=False)
pokemon.head()
pokemon = pokemon[pokemon['capture_rate'] != '30 (Meteorite)255 (Core)']
pokemon = pokemon.astype({
    'capture_rate': int
})


# This is the Complete Pokemon Dataset acquired from https://www.kaggle.com/rounakbanik/pokemon used in my project 2. It contains 801 observations and 41 variables.
```

```python
x = pokemon['height_m']
y = pokemon['capture_rate']
pokemon.plot.scatter(x='height_m',y='capture_rate', c = 'is_legendary',colormap='winter')
plt.title('Capture rate of pokemon vs height in meters')
plt.xlabel('Pokemon height in meters')
plt.ylabel('Capture rate of pokemon ')
plt.show()
# we can see that there is a weak negative correlation between pokemon height and capture rate. Taller pokemon tend to be more difficult to capture. 
# This may be explained by legendary pokemon being taller in general, and legendaries are by design harder to capture.
```

```python
legends = pokemon[pokemon['is_legendary']==1]
regs = pokemon[pokemon['is_legendary']==0]
print(legends['height_m'].mean())
print(regs['height_m'].mean())
# We see from this that the mean height of legendaries is more than twice that of regular pokemon.
```
