```python
# get the URL
year = '2019'
string = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

# format year to URL
url = string.format(year)
url
```

```python
years = [2015, 2016, 2017, 2018, 2019]
for year in years:
  url = string.format(year)
  print(url)
```

```python
#@title Default title text
import pandas as pd
```

```python
df = pd.read_html(url, header = 0)
df
```

```python
type(df[0])
```

```python
df2019 = df[0]
len(df2019)
```

```python
# table header is repeated multiple times
df2019[df2019.Age == 'Age']
```

```python
df = df2019.drop(df2019[df2019.Age == 'Age'].index)
df.shape
```

```python
import seaborn as sns
```

```python
sns.distplot(df.PTS,
             kde=False,
             hist_kws=dict(edgecolor='black', linewidth=2))
```

```python
# fill missing values with 0
df = df.fillna(0)
```

```python
df = pd.read_csv('nba2019.csv')
```

```python
df.dtypes
```

```python
# Which Player Scored the Most Points Per Game? (PTS)
df[df.PTS == df.PTS.max()]
```

```python
# Which player had the most 3-Point Field Goals per game?
df[df['3P'] == df['3P'].max()]
```

```python
# Which player had the highest Assists per game?
df[df['AST'] == df['AST'].max()]
```

```python
#GroupBy() function
# Which player scored the highest PTS in the LA Lakers??
LAL = df.groupby('Tm').get_group('LAL')
LAL
```

```python
# Of the 5 positions, which one scores the most points?
# First, group players by their positions
df.groupby('Pos').PTS.describe()
```

```python
positions = ['C', 'PF', 'SF', 'PG', 'SG']
POS = df[df['Pos'].isin(positions)]
POS.groupby('Pos').PTS.describe()
```

```python
PTS = df[['Pos','PTS']]
PTS = PTS[PTS['Pos'].isin(positions)]
PTS
```

```python
import matplotlib.pyplot as plt

g = sns.FacetGrid(PTS, col="Pos")
g.map(plt.hist,"PTS")
```

```python
sns.boxplot(x = 'Pos', y = 'PTS', data = PTS)
```

```python
# Heat Map
import numpy as np

corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
  f, ax = plt.subplots(figsize=(7,5))
  ax = sns.heatmap(corr, mask=mask, vmax=1, square=True)
```
